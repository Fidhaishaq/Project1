```python
from sklearn.preprocessing import LabelEncoder
```

```python
l=LabelEncoder() # label encoder is used to convert the alphatics to numeric values.
```

```python
k=l.fit_transform(d['mainroad'])
```

```python
d['mainroad']=l.fit_transform(d['mainroad']) #for converting it into numbers.
```

```python
d['guestroom']=l.fit_transform(d['guestroom'])
d.head()
```

```python
d['prefarea']=l.fit_transform(d['prefarea'])
d.head()
```

```python
d['furnishingstatus']=l.fit_transform(d['furnishingstatus'])
d.head()
```

```python
x=d[['area','bedrooms','stories','mainroad','guestroom','parking','prefarea','furnishingstatus']]
x.head()
```

```python
y=d[['price']] # price is storing into a vaariable named y.
y.head()
# here price is the target 
#so y is the target and x is the features.
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2) #0.2 denotes thee  20%.
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(xtrain,ytrain) 
model.predict(xtest)
model.score(xtest,ytest)
```
